```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df = pd.read_csv("../data/churn.csv")
```

```python
df = df.copy()

train, test = train_test_split(
    df,
    test_size=0.2,
    stratify=df["Churn"],
    random_state=42
)

train.to_csv("../data/train.csv", index=False)
test.to_csv("../data/test.csv", index=False)
```
